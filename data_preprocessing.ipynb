### Удаление упоминаний, хэштегов, ссылок, RT

```python
import re
```

```python
#упоминания
df['text2'] = [re.sub(r'@\w+', "", x) for x in df.text]
#ссылки
df['text2'] = [re.sub(r'http[^\s]+', "", x) for x in df.text2]
#хэштеги
df['text2'] = [re.sub(r'#\w+', "", x) for x in df.text2]
#RT
df['text2'] = [re.sub(r'RT +: +', "", x) for x in df.text2]
#\n
df['text2'] = [re.sub(r'\n', " ", x) for x in df.text2]
```

```python
#в нижний регистр
df['text'] = [x.lower() for x in df.text2]
df['text'] 
```

```python
df.drop(columns='text2',inplace=True)
```

```python
train = df[:int(len(df)*0.8)]
val = df[int(len(df)*0.8):int(len(df)*0.9)]
test = df[int(len(df)*0.9):]
```

```python
test_public = test.iloc[:, [0]]
```

```python
test_public
```

```python
train.to_csv("data/train.csv", header=["text", "class"])
val.to_csv("data/val.csv", header=["text", "class"])
test_public.to_csv("data/test.csv", header=["text"])
```
